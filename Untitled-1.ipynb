```python
import re
import matplotlib.pyplot as plt

# 1. 读取日志文件
with open("ckpts/new2/log.txt", "r") as file:
    log_content = file.read()

# 2. 解析loss值
# loss_values = [float(loss) for loss in re.findall(r"train_loss_w: ([\d\.]+)", log_content)]
loss_values = [ float(loss) for loss in re.findall(r'"train_loss_latents_b": ([\d\.]+)', log_content)]
loss_values2 = [ float(loss) for loss in re.findall(r'"train_loss_w": ([\d\.]+e[\-\+]\d+)', log_content)]


# 3. 绘制loss曲线
plt.plot(loss_values)
plt.title("Latent Loss Curve")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

plt.plot(loss_values2)
plt.title("Watermark Loss Curve")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

loss_values3 = [ float(loss) for loss in re.findall(r'"train_loss": ([\d\.]+)', log_content)]
plt.plot(loss_values3[300:])
plt.title("Loss Curve")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
loss_values4 = [ float(loss) for loss in re.findall(r'"train_lr": ([\d\.]+e[\-\+]\d+)', log_content)]

plt.plot(loss_values4)
plt.title("Learning Rate")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```
