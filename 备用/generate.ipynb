```python
import os
os.environ["HF_ENDPOINT"] = "https://hf-mirror.com"

os.environ['CUDA_VISIBLE_DEVICES']='2'

from diffusers import AutoPipelineForText2Image
from diffusers import AutoPipelineForImage2Image
from diffusers.utils import load_image, make_image_grid
import torch

# 以下代码引入SDXL模型
model= "/ssd-sata1/hqq/stable_signature/stable-diffusion-2-1-base"

pipeline_text2image = AutoPipelineForText2Image.from_pretrained(
         model, torch_dtype=torch.float16, variant="fp16", use_safetensors=True
).to("cuda")
```

```python
import PIL 

init_image=load_image('/ssd-sata1/hqq/stable_signature/WM encoder_decoder/img_outputs/init.png')

# 以下代码通过原始图像和提示词，通过图生图的方式生成新图像

prompt = ""
pipeline= AutoPipelineForImage2Image.from_pipe(pipeline_text2image)
image = pipeline(prompt, image=init_image, strength=0.8, guidance_scale=10.5).images[0]

image.save('/ssd-sata1/hqq/stable_signature/WM encoder_decoder/img_outputs/img2img.png')
```
